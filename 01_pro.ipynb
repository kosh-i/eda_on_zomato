```python
df =df.drop(columns=["Unnamed: 0", "Unnamed: 0.1"])
```

```python
df.columns =(
    df.columns
    .str.lower()
    .str.replace(" ","_")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

```python
df["rate_out_of_5"] = pd.to_numeric(df["rate_out_of_5"], errors="coerce")
```

```python
df["rate_out_of_5"].isnull().sum()
```

```python
df["avg_cost_two_people"] = pd.to_numeric(
    df["avg_cost_two_people"], errors="coerce"
)
```

```python
df["avg_cost_two_people"].fillna(
    df["avg_cost_two_people"].median(), inplace=True
)
```

```python
df["online_order"] = df["online_order"].map({"Yes": 1, "No": 0})
df["table_booking"] = df["table_booking"].map({"Yes": 1, "No": 0})
```

```python
df.duplicated().sum()
```

```python
df.info()
df.isnull().sum()
```

```python
df.to_csv("../data/cleaned/zomato_cleaned.csv", index=False)
```

```python
import matplotlib.pyplot as pyplot
import seaborn as sns 


plt.style.use("default")
```

```python
df["restaurant_type"].value_counts().head(30).plot(kind="bar")
plt.title("Top Restaurant Types on Zomato")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["rate_out_of_5"], bins=20)
plt.title("Rating Distribution")
plt.show()
```

```python
sns.boxplot(x=df["avg_cost_two_people"])
plt.title("Cost Distribution for Two People")
plt.show()
```

```python

sns.scatterplot(
    x="avg_cost_two_people",
    y="rate_out_of_5",
    data=df
)
plt.title("Cost vs Rating")
plt.show()
```

```python
sns.boxplot(
    x="online_order",
    y="rate_out_of_5",
    data=df
)
plt.title("Online Order Impact on Rating")
plt.show()
```

```python
sns.barplot(
    x="table_booking",
    y="avg_cost_two_people",
    data=df
)
plt.title("Table Booking vs Cost")
plt.show()
```

```python
df["area"].value_counts().head(10).plot(kind="bar")
plt.title("Top Areas with Most Restaurants")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
sns.heatmap(
    df[["rate_out_of_5", "num_of_ratings", "avg_cost_two_people"]].corr(),
    annot=True
)
plt.title("Correlation Heatmap")
plt.show()
```

### Key Insights
1. Quick Bites dominate the platform, indicating delivery-first behavior.
2. Most restaurants are rated between 3.0 and 4.0.
3. Higher cost does not guarantee higher ratings.
4. Restaurants offering online orders tend to have slightly better ratings.
5. Table booking is associated with higher-cost restaurants.
6. Certain areas have very high restaurant density.
